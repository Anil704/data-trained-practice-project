```python
x = df.drop(columns=['AveragePrice'],axis=1)
y = df['AveragePrice']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error,mean_absolute_error
from sklearn.model_selection import cross_val_score
```

```python
scaler = StandardScaler()
x_scaler = scaler.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_scaler,y,test_size = 0.29,random_state=38)
```

```python
lr=LinearRegression()
kn=KNeighborsRegressor()
rf=RandomForestRegressor()
ad=AdaBoostRegressor()
```

```python
lr.fit(x_train,y_train)
y_pred =lr.predict(x_test)
print('r2 score',r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
print(cross_val_score(lr,x,y,cv=5).mean())
```

```python
kn.fit(x_train,y_train)
y_pred = kn.predict(x_test)
print('r2 score',r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
print(cross_val_score(kn,x,y,cv=5).mean())
```

```python
rf.fit(x_train,y_train)
y_pred = rf.predict(x_test)
print('r2 score',r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
print(cross_val_score(rf,x,y,cv=5).mean())
```

```python
ad.fit(x_train,y_train)
y_pred = ad.predict(x_test)
print('r2 score',r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
print(cross_val_score(ad,x,y,cv=5).mean())
```
